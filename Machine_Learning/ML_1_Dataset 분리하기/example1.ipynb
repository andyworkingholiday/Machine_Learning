```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
data=np.loadtxt('iris.data', unpack=True, delimiter=',',skiprows=0)

iris=pd.read_csv("Iris.data")
iris.head()
```

```python
#1번문제//
k=1

for i in range(0,3):
    for j in range(i+1,4):
        plt.plot(data[i], data[j], 'o', markersize=3)
        if i==0:
            plt.xlabel('sepal length')
        elif i==1:
            plt.xlabel('sepal width')
        elif i==2:
            plt.xlabel('petal length')
        elif i==3:
            plt.xlabel('petal width')
            
        if j==1: 
            plt.ylabel('sepal width')
        elif j==2:
            plt.ylabel('petal legnth')
        elif j==3:
            plt.ylabel('petal.width')
        plt.grid()
        plt.suptitle('2D PLOT', fontsize=16)
        s=str(k)
        plt.savefig(s+'.png')
        k+=1
        ## 파일을 하나씩 출력하기 위해 정수 k를 하나씩 더해주었습니다.
        plt.show()
        
```
