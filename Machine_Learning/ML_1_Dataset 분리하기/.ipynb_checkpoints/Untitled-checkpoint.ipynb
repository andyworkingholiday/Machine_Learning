```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
data=np.loadtxt('iris.data', unpack=True, delimiter=',',skiprows=0)
```

```python
plt.plot(data[1], np.zeros_like(data[0]), 'o', markersize=3)
plt.xlabel('age')
plt.grid()
plt.suptitle('1D PLOT', fontsize=16)
plt.show()
```

```python
for i in range(0,3):
    for j in range(i+1,4):
        plt.plot(data[i], data[j], 'o', markersize=3)
        if i==0:
            plt.xlabel('sepal length')
        elif i==1:
            plt.xlabel('sepal width')
        elif i==2:
            plt.xlabel('petal length')
        elif i==3:
            plt.xlabel('petal width')
            
        if j==1: 
            plt.ylabel('sepal width')
        elif j==2:
            plt.ylabel('petal legnth')
        elif j==3:
            plt.ylabel('petal.width')
        plt.grid()
        plt.suptitle('2D PLOT', fontsize=16)
        plt.savefig('sex.png')
        plt.show()
        
```

```python
fig=plt.figure()
ax=Axes3D(fig)

X=data[2]
Y=data[3]
Z=data[4]

ax.scatter(X,Y,Z)

ax.set_xlabel('sepal length')
ax.set_ylabel('petal length')
ax.set_zlabel('class')
plt.suptitle('3d plotting', fontsize=16)
plt.show()
```
