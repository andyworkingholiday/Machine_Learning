```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import random

#X, Y   
x = np.random.randn(4, 2)
#theta=np.random.randn(3,1)
y = np.random.randn(4,1)
#cost_theta=np.dot(x,theta)-y


#cost function 
def cost_function(theta1, theta2, y):
    result=0
    for i in range(4):
        temp = 0
        temp = x[i][0]*theta1 + x[i][1]*theta2 - y[i]
        result += temp**2
    
    return result

#cost function  theta1  
def del_theta1(theta1,theta2,y):
    result=0
    for i in range(4):
        temp = 0
        temp = 2 * (x[i][0]*theta1 + x[i][1]*theta2 - y[i]) * x[i][0]
        result += temp
        
    return result

#cost function  theta2  
def del_theta2(theta1,theta2,y):
    result=0
    for i in range(4):
        temp = 0
        temp = 2 * (x[i][0]*theta1 + x[i][1]*theta2 - y[i]) * x[i][1]
        result += temp
        
    return result

#theta1, theta2    
def gradient_descent(first_theta1, first_theta2, y, iters, c):
    history_theta1 = [first_theta1]
    history_theta2 = [first_theta2]
    theta1 = first_theta1
    theta2 = first_theta2
    learning_rate = 0.005
    i=0
    while(i < iters):
        gradient1 = del_theta1(theta1, theta2, y)
        gradient2 = del_theta2(theta1, theta2, y)
        new_theta1 = theta1 - learning_rate*gradient1
        new_theta2 = theta2 - learning_rate*gradient2
        if (cost_function(new_theta1[0], new_theta2[0], y) < cost_function(theta1, theta2, y)):
            history_theta1.append(new_theta1[0])
            history_theta2.append(new_theta2[0])
            theta1 = new_theta1[0]
            theta2 = new_theta2[0]
            learning_rate *= 1.2
            i = i+1
           
        else:
            learning_rate *= 0.5
            print(i, "\t", "Calculating Learning late again")
            print(new_theta1[0], "\t", new_theta2[0])
            
        
    if (c==1):
        return history_theta1
    elif (c==2):
        return history_theta2

history1 = gradient_descent(2.8, -1.4, y, 30, 1)
history2 = gradient_descent(2.8, -1.4, y, 30, 2)
    
fig = plt.figure(figsize = (20,15))
ax = fig.gca(projection = '3d')
a = np.arange(-3, 3, 0.2)
b = np.arange(-3, 3, 0.2)
a, b = np.meshgrid(a, b)
outcome = cost_function(a, b, y)
surf = ax.plot_surface(a, b, outcome, rstride=1, cstride=1, alpha=0.3, linewidth=0, antialiased=False, cmap='rainbow')

a=np.array(history1)
b=np.array(history2)
outcome = cost_function(a, b, y)
ax.scatter(a, b, outcome, color="r");

print(outcome)
print("Final theta1 value is ", history1[30])
print("Final theta2 value is ", history2[30])

plt.xlabel('Theta1')
plt.ylabel('Theta2')

plt.show()
        


```
